```python
def min_coin_number(c, l_coins):
    """
    Devuelva el minimo numero de monedas de l_coin necesarias para dar cambio de una cantidad c.
    n(i,v) = {min (n(i-1,v) si no Ci); 1 + n(i,v-Ci) si Ci}
    """
    m = {i: {j:0 for j in range(c+1)} for i in range(len(l_coins))}
    
    for i in range(1, c+1): 
        m[0][i] = i
        
    for i in range(1, len(l_coins)):
        m[i][0] = 0 
        
    for i in range(1, len(l_coins)):
        for j in range(1, c+1):
            res = j-l_coins[i]
            if (res >= 0): 
                m[i][j] = min(m[i-1][j], 1+m[i][j-l_coins[i]])
            else:
                m[i][j] = m[i-1][j]
    return m, m[i][j]
```

```python
def optimal_change(c, l_coins):
    """
    Devuelve un dict con las monedas de cada tipo en l_coin necesarias para dar cambio de una cantidad c.
    
    """
    l = len(l_coins)
    m = {i: {j:0 for j in range(l)} for i in range(c+1)}
    
    for i in range(l):
        m[0][i] = 1
        
    for i in range(1, c+1):
        for j in range(l):
            if (i-l_coins[j] >= 0):
                res1 = m[i-l_coins[j]][j]
            else:
                res1 = 0
            if(j>=1):
                res2 = m[i][j-1]
            else:
                res2 = 0
            m[i][j] = res1 + res2
    return m
```

```python
l2 = [1,2,3]
print (optimal_change(4, l2))
```

```python
def max_matrix_common_subsequence(str_1, str_2):
    l1, l2 = len(str_1)+1, len(str_2)+1
    m = [[None for j in range(l1)] for i in range(l2)]
    
    m[0][0] = 0
    for i in range(1, l1):
        m[0][i] = 0
    for i in range(1, l2):
        m[i][0] = 0

    for i in range(1, l2):
        for j in range(1, l1):
            if str_2[i-1] == str_1[j-1]:
                m[i][j] = m[i-1][j-1] + 1
            else:
                m[i][j] = max(m[i-1][j], m[i][j-1])            
    return m


def max_length_common_subsequence(str_1, str_2):
    j, i = len(str_1), len(str_2)
    m = max_matrix_common_subsequence(str_1, str_2)
    print(m)
    return m[i][j]
```

```python
def find_max_common_subsequence(str_1, str_2):
    i, j = len(str_1)+1, len(str_2)+1
    m = max_matrix_common_subsequence(str_1, str_2)
    inds = max_length_common_subsequence(str_1, str_2)
    print(m,inds)
    aux = ""
    while (i > 0 and j > 0):
        if str_1[i-1] == str_2[i-1]:
            aux[inds-1] = str_1[i-1]
            i = i-1
            j = j-1
            inds = inds-1
        elif m[i-1][j] > m[i][j-1]:
            i = i-1
        else:
            j = j-1
    return aux
```

```python
def optimal_order(l_probs):
    length = len(l_probs)
    keys = range(1, length+1)
    aux = {j:{i:0 for i in range(length)} for j in keys}
    
    for l in keys:
        for i in range(length-l+1):
            j = i+l-1
            aux[i][j] = np.inf
            for k in range(i,j+1):
                if (k>i):
                    c = aux[i][k-1]
                else:
                    c = 0
                if (k<j):
                    c = c + aux[k+1][j]
                c = c + sum(l_probs[i][j])
                if (c < aux[i][j]):
                    aux[i][j] = c
    return aux[0][length-1]
```

```python
def optimal_change2(c, l_coins):
    aux = {i:0 for i in range(c+1)} 
   
    aux[0] = 1
  
    for i in range(0,len(l_coins)): 
        for j in range(l_coins[i],c+1): 
            aux[j] += aux[j-l_coins[i]] 
  
    return aux[c]
```
