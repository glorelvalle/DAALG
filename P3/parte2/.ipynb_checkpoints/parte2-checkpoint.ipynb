```python
import numpy as np
import sys
```

```python
def min_coin_number(c, l_coins):
    """
    Devuelva el minimo numero de monedas de l_coin necesarias para dar cambio de una cantidad c.
    n(i,v) = {min (n(i-1,v) si no Ci); 1 + n(i,v-Ci) si Ci}
    """
    # ini
    aux = [[0 for i in range(c+1)] for j in range(len(l_coins))]
    for i in range(1, c+1): 
        aux[0][i] = sys.maxsize
    # caso base
    for i in range(1, len(l_coins)):
        aux[i][0] = 0
    
    for i in range(1, len(l_coins)):
        for j in range(1, c+1):
            res = j-l_coins[i]
            if (res < sys.maxsize):
                if (res >= 0): 
                    aux[i][j] = min(aux[i-1][j], 1+aux[i][j-l_coins[i]])
                else:
                    aux[i][j] = aux[i-1][j]
    print(aux)
    return aux[i][j]
```

```python
def optimal_change(c, l_coins):
    """
    Devuelve un dict con las monedas de cada tipo en l_coin necesarias para dar cambio de una cantidad c.
    
    """
    # ini
    l = len(l_coins)
    aux = [[0 for i in range(l)] for j in range(c+1)] 
    #caso base
    for i in range(l):
        aux[0][i] = 1
    for i in range(1, c+1):
        for j in range(l):
            if (i-l_coins[j] >= 0):
                res1 = aux[i-l_coins[j]][j]
            else:
                res1 = 0
            if(j>=1):
                res2 = aux[i][j-1]
            else:
                res2 = 0
            aux[i][j] = res1 + res2
    return aux    
```

```python
l2 = [1,2,3]
print (optimal_change(4, l2))
```
