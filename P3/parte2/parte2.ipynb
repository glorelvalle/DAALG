```python
def min_coin_number(c, l_coins):
    """
    Devuelva el minimo numero de monedas de l_coin necesarias para dar cambio de una cantidad c.
    n(i,v) = {min (n(i-1,v) si no Ci); 1 + n(i,v-Ci) si Ci}
    """
    m = {i: {j:0 for j in range(c+1)} for i in range(len(l_coins))}
    
    for i in range(1, c+1): 
        m[0][i] = i
        
    for i in range(1, len(l_coins)):
        m[i][0] = 0 
        
    for i in range(1, len(l_coins)):
        for j in range(1, c+1):
            res = j-l_coins[i]
            if (res >= 0): 
                m[i][j] = min(m[i-1][j], 1+m[i][j-l_coins[i]])
            else:
                m[i][j] = m[i-1][j]
    return m, m[i][j]
```

```python
def optimal_change(c, l_coins):
    """
    Devuelve un dict con las monedas de cada tipo en l_coin necesarias para dar cambio de una cantidad c.
    
    """
    l = len(l_coins)
    m = {i: {j:0 for j in range(l)} for i in range(c+1)}
    
    for i in range(l):
        m[0][i] = 1
        
    for i in range(1, c+1):
        for j in range(l):
            if (i-l_coins[j] >= 0):
                res1 = m[i-l_coins[j]][j]
            else:
                res1 = 0
            if(j>=1):
                res2 = m[i][j-1]
            else:
                res2 = 0
            m[i][j] = res1 + res2
    return m
```

```python
def max_matrix_common_subsequence(str_1, str_2):
    l1, l2 = len(str_1)+1, len(str_2)+1
    m = [[0 for j in range(l2)] for i in range(l1)]

    for i in range(1, l1):
        for j in range(1, l2):
            if i == 0 or j == 0:
                m[i][j] = 0
            elif str_1[i-1] == str_2[j-1]:
                m[i][j] = m[i-1][j-1] + 1
            else:
                m[i][j] = max(m[i-1][j], m[i][j-1])            
    return m


def max_length_common_subsequence(str_1, str_2):
    i, j = len(str_1), len(str_2)
    m = max_matrix_common_subsequence(str_1, str_2)
    print(m)
    return m[i][j]
```
