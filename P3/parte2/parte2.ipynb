```python
def min_coin_number(c, l_coins):
    """
    Devuelva el minimo numero de monedas de l_coin necesarias para dar cambio de una cantidad c.
    m = len(l_coins)
    If c == 0, then 0 coins required.
    If c > 0
       min_coin_number(coins[0..m-1], c) = min {1 + min_coin_number(c-l_coins[i])} 
                               where i varies from 0 to m-1 
                               and l_coins[i] <= c 
    """
    # ini
    aux = [0 for i in range(c+1)]
    print (aux)
    
    # caso base
    aux[0] = 0
    
    for i in range(1, c + 1): 
        aux[i] = sys.maxsize
   
    for i in range(1, c+1):   
        for j in range(len(l_coins)): 
            if (l_coins[j] <= i): 
                res = aux[i - l_coins[j]]
                print (res)
                if (res != sys.maxsize and res+1 < aux[i]): 
                    aux[i] = res + 1
    print(aux)
    return aux[c]
```

```python
def optimal_change(c, l_coins):
    """
    Devuelve un dict con las monedas de cada tipo en l_coin necesarias para dar cambio de una cantidad c.
    """
    aux = [0 for i in range(c+1)]
    aux[0] = 1
    l = len(l_coins)
    print(aux)
    for i in range(1, c+1): 
        for j in range(l):
            if(i-l_coins[j] >= 0):
                res1 = aux[i-l_coins[j]][j]
            elif(j>=1):
                res2 = aux[i][j-1]
            else:
                res1 = 0
                res2 = 0
            aux[i][j] = res1 + res2
    return aux[c][l-1] 
```
