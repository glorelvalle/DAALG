```python
def min_coin_number(c, l_coins):
    """
    Devuelva el minimo numero de monedas de l_coin necesarias para dar cambio de una cantidad c.
    n(i,v) = {min (n(i-1,v) si no Ci); 1 + n(i,v-Ci) si Ci}
    """
    aux = {i: {j:0 for j in range(c+1)} for i in range(len(l_coins))}
    
    for i in range(1, c+1): 
        aux[0][i] = i
        
    for i in range(1, len(l_coins)):
        aux[i][0] = 0 
        
    for i in range(1, len(l_coins)):
        for j in range(1, c+1):
            res = j-l_coins[i]
            if (res >= 0): 
                aux[i][j] = min(aux[i-1][j], 1+aux[i][j-l_coins[i]])
            else:
                aux[i][j] = aux[i-1][j]
    return aux, aux[i][j]
```

```python
def optimal_change(c, l_coins):
    """
    Devuelve un dict con las monedas de cada tipo en l_coin necesarias para dar cambio de una cantidad c.
    
    """
    l = len(l_coins)
    aux = {i: {j:0 for j in range(l)} for i in range(c+1)}
    
    for i in range(l):
        aux[0][i] = 1
        
    for i in range(1, c+1):
        for j in range(l):
            if (i-l_coins[j] >= 0):
                res1 = aux[i-l_coins[j]][j]
            else:
                res1 = 0
            if(j>=1):
                res2 = aux[i][j-1]
            else:
                res2 = 0
            aux[i][j] = res1 + res2
    return aux
```

```python
def max_length_common_subsequence(str_1, str_2):
    l1, l2 = len(str_1)+1, len(str_2)+1
    aux = {i: {j:None for j in range(l1)} for i in range(l2)}
    
    aux[0][0] = 0
    for i in range(1, l1):
        aux[0][i] = 0
    for i in range(1, l2):
        aux[i][0] = 0

    for i in range(1, l2):
        for j in range(1, l1):
            if str_2[i-1] == str_1[j-1]:
                aux[i][j] = aux[i-1][j-1] + 1
            else:
                aux[i][j] = max(aux[i-1][j], aux[i][j-1])            
    return aux, aux[i][j]
```

```python
def optimal_order(l_probs):
    l = len(l_probs)
    keys = range(1, len(l_probs)+1)
    aux = {{i:0 for i in range(l) for j in keys}}
    for i in range(l):
        aux[i][i] = l_probs[i]
```
